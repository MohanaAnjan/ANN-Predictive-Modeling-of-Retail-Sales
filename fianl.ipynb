```python

x=['Store', 'Temperature', 'Fuel_Price',  'CPI', 'Unemployment',
        'Dept', 'Weekly_Sales', 'Type', 'Size', 'year', 'month',
       'day']
for i in x:
    plt.figure(figsize=(8,8),dpi=100)
    plt.boxplot(df1[i])  # Remove the extra brackets
    plt.title(i)  # set the title of the plot to the column name
    plt.show()
```

REMOVAL OF OUTLIERS FOR UNEMPLOYMENT COLUMN

```python
df1['Unemployment'] =df1['Unemployment'].clip(lower=df1['Unemployment'].quantile(0.05),upper=df1['Unemployment'].quantile(0.85))
```
